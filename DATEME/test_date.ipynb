```python
import os

images_folder = r'../sandboxACK/images_12'  # 注意用 r'' 防止转义
def rename_folder_files(folder, prefix_file='image'):
    files = os.listdir(folder)
    
    # 例如按序号重命名所有图片
    for i, filename in enumerate(files, start=1):
        # 获取文件扩展名
        ext = os.path.splitext(filename)[1]  # 包含点，如 '.jpg'
        
        # 构造新文件名
        new_name = f"{prefix_file}_{i:03d}{ext}"  # image_001.jpg, image_002.jpg ...
        # 拼接完整路径
        old_path = os.path.join(images_folder, filename)
        new_path = os.path.join(images_folder, new_name)
        # 重命名
        os.rename(old_path, new_path)
    print(f'[SUCCES] files in folder renamed: {prefix_file}_idx{ext}')
    return 

rename_folder_files(images_folder, prefix_file='img')
```

```python

from PIL import Image
import os
import re

# ======================
# 参数区
# ======================
rows = 3
cols = 4
folder = r"../sandboxACK/output_images_16_2"
output_path =f "calendar_overview_{rows}x{cols}_black.png"



img_w = 480      # 单张宽度
img_h = 270      # 单张高度（16:9）
gap = 32         # 图片间距

bg_color = "#000000"   # 黑色背景

# ======================
# 读取图片并按月份排序
# ======================

def extract_month(filename):
    # 从类似 '0_calender_2026_1.png' 提取最后的月份数字
    match = re.search(r'_(\d+)\.png$', filename)
    return int(match.group(1)) if match else 0

images = [
    os.path.join(folder, f)
    for f in os.listdir(folder)
    if f.lower().endswith(".png")
]

# 按月份数字排序
images.sort(key=lambda x: extract_month(os.path.basename(x)))

assert len(images) >= 12, "图片数量不足 12 张"
images = images[:12]

# ======================
# 计算画布尺寸
# ======================

canvas_w = cols * img_w + (cols - 1) * gap
canvas_h = rows * img_h + (rows - 1) * gap

canvas = Image.new("RGB", (canvas_w, canvas_h), bg_color)

# ======================
# 排版贴图（左到右，从上到下）
# ======================

for idx, path in enumerate(images):
    img = Image.open(path).convert("RGB")
    img = img.resize((img_w, img_h), Image.LANCZOS)

    row = idx // cols
    col = idx % cols

    x = col * (img_w + gap)
    y = row * (img_h + gap)

    canvas.paste(img, (x, y))

# ======================
# 保存
# ======================

canvas.save(output_path, quality=95)
print(f"✅ 总览图已生成：{output_path}")
```
