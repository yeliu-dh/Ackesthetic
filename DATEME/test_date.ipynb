```python
images_folder='sandboxDATE\images_16'
files=os.listdir(images_folder)

# 例如按序号重命名所有图片
for i, filename in enumerate(files, start=1):
    # 获取文件扩展名
    ext = os.path.splitext(filename)[1]  # 包含点，如 '.jpg'
    # 构造新文件名
    new_name = f"image_{i:03d}{ext}"  # image_001.jpg, image_002.jpg ...
    # 拼接完整路径
    old_path = os.path.join(images_folder, filename)
    new_path = os.path.join(images_folder, new_name)
    # 重命名
    os.rename(old_path, new_path)
```
